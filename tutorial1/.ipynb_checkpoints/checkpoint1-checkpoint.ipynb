<h2> Data Cleanup </h2>

Import `numpy`, `pandas`, and `matplotlib`.

(Feel free to import other libraries!)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Load in the "states_edu.csv" dataset and take a look at the head of the data

```python
df = pd.read_csv("../data/states_edu.csv")
df.head()
```

You should always familiarize yourself with what each column in the dataframe represents. Read about the states_edu dataset here: https://www.kaggle.com/noriuk/us-education-datasets-unification-project

Use this space to rename columns, deal with missing data, etc. _(optional)_

```python
df.rename({
    "GRADES_PK_G":"ENROLL_PK",
    "GRADES_KG_G":"ENROLL_KINDER",
    "GRADES_4_G":"ENROLL_4",
    "GRADES_8_G":"ENROLL_8",
    "GRADES_12_G":"ENROLL_12",
    "GRADES_1_8_G":"ENROLL_PRIMARY",
    "GRADES_9_12_G":"ENROLL_HS",
    "GRADES_ALL_G":"ENROLL_ALL",
    "ENROLL":"ENROLL_ALL_EST"
    },
    axis=1, inplace=True)
df.dropna(subset=["AVG_MATH_4_SCORE"], inplace=True)
df.shape
```

<h2> Feature Engineering </h2>

After exploring the data, you can choose to modify features that you would use to predict the performance of the students on your chosen response variable. 

You can also create your own features. For example, perhaps you figured that maybe a state's expenditure per student may affect their overall academic performance so you create a expenditure_per_student feature.

Use this space to modify or create features.

```python
df["SUPPORT_EXP_PER_STUDENT"] = df["SUPPORT_SERVICES_EXPENDITURE"] / df["ENROLL_ALL"]
df["SUPPORT_EXP_PER_STUDENT"].describe()
```

Feature engineering justification: I figured that since 4th grade math is where students are likely beginning to figure out what subjects they like/are good at - and thus may begin to need extra help, so knowing how much each state spent on average for support services could have a correlation for the math performance of their respective students

Investigate the relationship between your chosen response variable and at least two predictors using visualizations. Write down your observations.

<h2> Data Creation </h2>

_Use this space to create train/test data_

```python
from sklearn.model_selection import train_test_split
```

```python
X = df[['SUPPORT_EXP_PER_STUDENT', 'AVG_READING_4_SCORE']].dropna()
y = df.loc[X.index]['AVG_MATH_4_SCORE']
y.fillna(y.median(), inplace=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3, random_state=42)
```

<h2> Prediction </h2>

ML Models [Resource](https://medium.com/@vijaya.beeravalli/comparison-of-machine-learning-classification-models-for-credit-card-default-data-c3cf805c9a5a)

```python
from sklearn.linear_model import LinearRegression
```

```python
# create your model here
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

## Evaluation

Choose some metrics to evaluate the performance of your model, some of them are mentioned in the tutorial.

```python
print(model.score(X_test, y_test))
np.mean(model.predict(X_test)-y_test)
```

We have copied over the graphs that visualize the model's performance on the training and testing set. 

Change `col_name` and modify the call to `plt.ylabel()` to isolate how a single predictor affects the model.

```python
col_name = 'SUPPORT_EXP_PER_STUDENT'

f = plt.figure(figsize=(12,6))
plt.scatter(X_train[col_name], y_train, color = "red")
plt.scatter(X_train[col_name], model.predict(X_train), color = "green")

plt.legend(['True Training','Predicted Training'])
plt.xlabel(col_name)
plt.ylabel('AVG_MATH_4_SCORE')
plt.title("Model Behavior On Training Set")
```

```python
col_name = 'SUPPORT_EXP_PER_STUDENT'

f = plt.figure(figsize=(12,6))
plt.scatter(X_test[col_name], y_test, color = "blue")
plt.scatter(X_test[col_name], model.predict(X_test), color = "black")

plt.legend(['True testing','Predicted testing'])
plt.xlabel(col_name)
plt.ylabel('AVG_MATH_4_SCORE')
plt.title("Model Behavior on Testing Set")
```

<h2> Summary </h2>

I chose to analyze the average math score for fourth graders by state and year. In the exploratory data anaylsis, I found that the data was tightly bunched and left skewed. I did two scatterplot visualizations with two diferent predictors. One was average readings core for fourth graders because I thought that performance in another school subject in the same age range would be a strong predictor of math performance. It turned out to be pretty strongly correlated, so I decided to use it. The other was a column that I created myself - average support expenditure per student - which I created by dividing support expenditure by total enrollment. I figured that average support expenditure would have a correlation with math scores because it shows which states devote resources towards tutoring and extra help. Average support expenditure turned out to have a weaker correlation than average 4th grade reading score, but I still wanted to try using it. Since the response variable was left skewed, I filled in missing values with the median. My model had an R^2 value of 0.68 and a mean error of 0.96, so it wasn't as strong of a model as the tutorial example. In the end, the model worked fairly well on the training set and not quite as well on the testing set, so it's possible there was some overfitting occurring.
